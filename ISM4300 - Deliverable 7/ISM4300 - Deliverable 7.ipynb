```python
#Author - Pedro Nader
#Deliverable 7 Python

#Importing Pandas
import pandas as pd

#Similar to what we did in C#. Create the loop that will flag when done.
done = False
#Loop that will keep running before we flag it as done
while done == False:
    #Use try except to avoid entering a string or decimal
    try:
        n = int(input('Input a whole number larger than 0: '))
        if n > 0 :
            done = True #Validate if true 
        else:
            print('Please Input a valid number. it has to be positive.') #Show error message if number is negative
    except:
        print('Do not enter a string or decimal') #message if user does not input a number or inputs a decimal

    
```

```python
#Create all lists
even = []
odd = []
fibo = [0,1]

#Loop trhough all the numbers
for i in range(int(n)):
    #append even and odd numbers to the list
    even.append(i*2)
    odd.append(i*2+1)

    #when the loop reaches the index 2, it will start adding numbres to the fibonacci sequence
    if i > 1:
        fibo.append(fibo[i-1] + fibo[1-2])

#Making sure they have the same length
print(len(even) == len(odd) == len(fibo))
```

```python
df = pd.DataFrame()
df['even'] = even
df['odd'] = odd
df['Fibo'] = fibo

df
```
